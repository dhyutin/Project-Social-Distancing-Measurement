```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Opens the inbuilt camera of laptop to capture video.
cap = cv2.VideoCapture("town.avi")
# i = 0
frames = [] 
while(cap.isOpened()):
    ret, frame = cap.read()
     
    # This condition prevents from infinite looping
    # incase video ends.
    if ret == False:
        break
    frame = cv2.resize(frame, (640, 480)) 
    #Save Frame by Frame into disk using imwrite method
    #cv2.imwrite('Frame'+str(i)+'.jpg', frame)
    frames.append(frame)
#     i += 1
cap.release()
cv2.destroyAllWindows()
```

```python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

img = frames[0]

frame = cv2.resize(img, (640, 480))
# using a greyscale picture, also for faster detection
#gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    
    
boxes, weights = hog.detectMultiScale(frame, winStride=(8,8) )
    
boxes = np.array([[x, y, x + w, y + h] for (x, y, w, h) in boxes])
    
for (xA, yA, xB, yB) in boxes:
# display the detected boxes in the colour picture
    cv2.rectangle(frames[0], (xA, yA), (xB, yB),(0, 255, 0), 2)
```

```python
# (xA, yA) is the starting point of a rectangle
# (xB, yB) is the ending point of a rectangle

import math

def calc_centroid(x1, y1, x2, y2):
    x = (x1 + x2)/2
    y = (y1 + y2)/2
    return (x, y)


def dist_euclidean(boxes, img, factor, xA, yA, xB, yB, threshold):
    centroids = []
    for i in boxes:
        value = calc_centroid(i[0], i[1], i[2], i[3])
        centroids.append(value)
        
    # Find Euclidean distance between all centroids
    for j in centroids:
        for k in centroids:
            if(j != k):
                a = np.array(j)
                b = np.array(k)
                dist = np.linalg.norm(a - b)
                print(dist)
                dist = dist * factor
                if(dist < threshold):
                    cv2.rectangle(frame, (xA, yA), (xB, yB),(255, 0, 0), 2)
                    cv2.line(img, (int(a[0]),int(a[1])), (int(b[0]),int(b[1])), (255, 0, 0), 1)
                else :
                    cv2.rectangle(frame, (xA, yA), (xB, yB),(0, 255, 0), 2)
                    cv2.line(img, (int(a[0]),int(a[1])), (int(b[0]),int(b[1])), (0, 255, 0), 1)
                
def y_parameter(factor, select):
    
    #2^y
    if(select == 1):
        return 2**factor
    
    #log2(y)
    elif(select == 2):
        return math.log(factor, 2)
    
    
threshold = 300
temp = []
for f in frames[:500]:    
    hog = cv2.HOGDescriptor()
    hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
    
    frame = cv2.resize(f, (640, 480))
        # using a greyscale picture, also for faster detection
    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    
    
    boxes, weights = hog.detectMultiScale(frame, winStride=(8,8) )
    
    boxes = np.array([[x, y, x + w, y + h] for (x, y, w, h) in boxes])
    
    for (xA, yA, xB, yB) in boxes:
            # display the detected boxes in the colour picture
#         cv2.rectangle(frame, (xA, yA), (xB, yB),(0, 255, 0), 2)
        factor = (yA + yB)/2
        factor = y_parameter(factor, 1)
        dist_euclidean(boxes, frame, factor, xA, yA, xB, yB)
    temp.append(frame)
    #plt.imshow(frame,cmap="gray")
```
